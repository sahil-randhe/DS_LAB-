```python
def perform_arithmetic_operations(num1, num2):
    """Performs basic arithmetic operations on two numbers."""
    print(f"Numbers: {num1}, {num2}")
    print(f"Addition: {num1} + {num2} = {num1 + num2}")
    print(f"Subtraction: {num1} - {num2} = {num1 - num2}")
    print(f"Multiplication: {num1} * {num2} = {num1 * num2}")

    if num2 != 0:
        print(f"Division: {num1} / {num2} = {num1 / num2}")
        print(f"Floor Division: {num1} // {num2} = {num1 // num2}")
        print(f"Modulus (Remainder): {num1} % {num2} = {num1 % num2}")
    else:
        print("Division by zero is not allowed.")

    print(f"Exponentiation: {num1} ** {num2} = {num1 ** num2}")

# Example usage:
perform_arithmetic_operations(10, 5)
print("\n---")
perform_arithmetic_operations(7, 3)
print("\n---")
perform_arithmetic_operations(15, 0)
```

This code defines a function `perform_arithmetic_operations` that takes two numbers as input and calculates their sum, difference, product, quotient (both float and integer division), remainder, and exponentiation. It also includes a check to prevent division by zero.
