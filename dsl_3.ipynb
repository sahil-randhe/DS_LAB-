```python
print("Enter the elements of the 3x3 matrix:")

matrix = []
for i in range(3):
    row = list(map(float, input(f"Enter row {i+1} (3 numbers separated by space): ").split()))
    matrix.append(row)

a, b, c = matrix[0]
d, e, f = matrix[1]
g, h, i = matrix[2]

# Determinant of the matrix
det = a*(e*i - f*h) - b*(d*i - f*g) + c*(d*h - e*g)

print("\nMatrix:")
for row in matrix:
    print(row)

if det == 0:
    print("\nInverse does not exist (determinant is 0).")
else:
    # Cofactor matrix
    C = [
        [(e*i - f*h), -(d*i - f*g), (d*h - e*g)],
        [-(b*i - c*h), (a*i - c*g), -(a*h - b*g)],
        [(b*f - c*e), -(a*f - c*d), (a*e - b*d)]
    ]

    # Adjoint (transpose of cofactor matrix)
    adjoint = [
        [C[0][0], C[1][0], C[2][0]],
        [C[0][1], C[1][1], C[2][1]],
        [C[0][2], C[1][2], C[2][2]]
    ]

    # Inverse = (1/det) * adjoint
    inverse = [[adjoint[r][c] / det for c in range(3)] for r in range(3)]

    print("\nInverse of the matrix:")
    for row in inverse:
        print(row)
```

```python


# Define the coefficient matrix A
A = np.array([[2, 3], [4, 5]])

# Define the constant vector B
B = np.array([10, 20])

# Solve the system of linear equations
x_and_y = np.linalg.solve(A, B)

print("The solution for x and y is:")
print(f"x = {x_and_y[0]:.2f}")
print(f"y = {x_and_y[1]:.2f}")
```
